###Q1. Explain GET and POST methods.
* The GET method is used to retrieve data from the server. When a client sends a GET request to a server, it includes a query string in the URL. This query string contains key-value pairs that specify the data that the client wants to retrieve. The server then responds with the requested data, which is usually in the form of an HTML page, a JSON object, or a file. GET requests are typically used for simple queries that don't modify data on the server.

* The POST method is used to submit data to a server. When a client sends a POST request, it includes a request body that contains the data that the client wants to submit. This data can be in any format, such as JSON, XML, or plain text. The server then processes the data and returns a response, which can be an HTML page, a JSON object, or a file. POST requests are typically used for more complex operations that modify data on the server, such as submitting a form or uploading a file.

####In summary, GET is used to retrieve data from a server, while POST is used to submit data to a server. Both methods are important for building web applications that interact with servers.




###Q2. Why is request used in Flask?

* In Flask, the request object is used to handle incoming HTTP requests from a client. When a client makes an HTTP request to a Flask web application, the request object is used to access data associated with the request, such as the request method, headers, form data, and query parameters.

* The request object is particularly useful for handling form data submitted by a client. When a client submits a form using a POST request, the request object can be used to access the form data and process it on the server side. The request object also allows the server to differentiate between different HTTP methods, such as GET and POST, and to handle each method accordingly.

* In addition to handling incoming requests, the request object can also be used to store data that is shared between different parts of a Flask application. For example, data can be stored in the request object and accessed by different views or templates.

####Overall, the request object is a fundamental part of the Flask framework that allows developers to build web applications that can handle incoming HTTP requests and respond accordingly.

###Q3. Why is redirect() used in Flask?

####In Flask, the redirect() function is used to redirect a client to a different URL. This function is useful for handling situations where a user needs to be redirected to a different page or resource within a Flask application.

* The redirect() function takes a URL as its argument and returns a response object that instructs the client to redirect to that URL. When a client receives a redirect response, it will automatically make a new request to the URL specified in the response.

####There are several situations where the redirect() function can be used in Flask:

* Redirecting after a form submission: When a user submits a form, the redirect() function can be used to redirect the user to a different page, such as a confirmation page or a thank you page.

* Handling authentication: If a user attempts to access a protected resource without being authenticated, the redirect() function can be used to redirect the user to a login page.

* Handling errors: If an error occurs during the processing of a request, the redirect() function can be used to redirect the user to an error page or a page that provides additional information about the error.

####Overall, the redirect() function is an important tool for handling redirects within Flask applications, and it can be used in a variety of situations to provide a better user experience and improve the overall functionality of a web application.

###Q4. What are templates in Flask? Why is the render_template() function used?

####In Flask, templates are used to generate dynamic HTML pages that can be customized based on the data provided by the application. Templates allow developers to separate the presentation layer from the application logic, making it easier to manage the code and modify the look and feel of the application.

####A template in Flask is a file that contains a combination of HTML, CSS, and Python code. The Python code is enclosed in special tags, which are used to generate dynamic content based on the data provided by the application.

####The render_template() function is used in Flask to render a template and generate an HTML response that can be sent back to the client. This function takes the name of a template as its argument and any additional data that needs to be passed to the template.

####When a template is rendered using the render_template() function, Flask will execute the Python code within the template and generate an HTML page that can be sent to the client. The data passed to the template can be used to generate dynamic content, such as a list of items or a table of data.

#### Overall, templates are an important part of building Flask applications, as they allow developers to create dynamic HTML pages that can be customized based on the data provided by the application. The render_template() function is a key tool for rendering templates and generating HTML responses in Flask.

##Q5. Create a simple API. Use Postman to test it.

```python
##app.py code
from flask import Flask,request ,render_template , jsonify

app = Flask(__name__)


@app.route('/')
def home_page():
    return render_template('index.html')


@app.route('/postman_action',methods=['POST'])
def math_ops1():
    if(request.method == 'POST'):
        ops = request.json['operation']
        num1 = int(request.json['num1'])
        num2 = int(request.json['num2'])
        if ops == 'add':
            r = num1+num2
            result = "The sum of " + str(num1) + ' and ' + str(num2) + " is " + str(r)
        if ops == 'subtract':
            r = num1-num2
            result = "The subtract of " + str(num1) + ' and ' + str(num2) + " is " + str(r)
        if ops == 'multiply':
            r = num1*num2
            result = "The multiply of " + str(num1) + ' and ' + str(num2) + " is " + str(r)
        if ops == 'divide':
            r = num1/num2
            result = "The divide of " + str(num1) + ' and ' + str(num2) + " is " + str(r)
            
        return jsonify(result)

if __name__=="__main__":
    app.run(host="0.0.0.0")
```
